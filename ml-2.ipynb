```python
import pandas as pd

# Creating a simple Series
data = [10, 20, 30, 40, 50]
series = pd.Series(data)

print("Pandas Series:")
print(series)
```

```python
print("Indices of Series:", series.index)
print("Values of Series:", series.values)
```

```python
import numpy as np

# Create a NumPy array
array = np.array([10, 20, 30, 40, 50])

print("NumPy Array:", array)
print("Pandas Series:\n", series)

# Compare properties
print("\nType of array:", type(array))
print("Type of series:", type(series))

# Both can be indexed, but Series has labels
print("\nAccess element [2]:")
print("NumPy Array:", array[2])
print("Pandas Series:", series[2])
```

```python
# Defining custom indices
students = pd.Series([89, 92, 76, 95], index=['Aman', 'Bhavna', 'Chetan', 'Diya'])

print("Series with Custom Indices:")
print(students)

# Access using index name
print("\nMarks of Diya:", students['Diya'])
```

```python
# Access by index number
print("2nd element (by position):", series[1])

# Access by index label (custom Series)
print("Marks of Aman:", students['Aman'])
```

```python
# Manual creation
data = {
    'Name': ['Aman', 'Bhavna', 'Chetan', 'Diya'],
    'Marks': [89, 92, 76, 95],
    'Subject': ['Math', 'Science', 'English', 'Math']
}

df = pd.DataFrame(data)
print("DataFrame created manually:")
print(df)
```

```python
import matplotlib.pyplot as plt

# Simple Line Plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y, color='blue', marker='o', linestyle='--')
plt.title("Simple Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()

# Bar Plot
students = ['Aman', 'Bhavna', 'Chetan', 'Diya']
marks = [89, 92, 76, 95]
plt.bar(students, marks, color='green')
plt.title("Marks of Students")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()

# Histogram
import numpy as np
data = np.random.randn(1000)
plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Pie Chart
sizes = [25, 35, 20, 20]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart Example")
plt.show()

# Scatter Plot
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y, color='purple')
plt.title("Scatter Plot Example")
plt.xlabel("Random X")
plt.ylabel("Random Y")
plt.show()
```
