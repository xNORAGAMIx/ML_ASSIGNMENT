```python
# 3) i
import pandas as pd

# Create a DataFrame manually using a dictionary
data = {
    'Name': ['Aman', 'Bhavna', 'Chetan', 'Diya', 'Esha'],
    'Age': [21, 20, 23, 22, 24],
    'Marks': [85, 92, 76, 88, 95],
    'City': ['Bhopal', 'Delhi', 'Pune', 'Bhopal', 'Chennai']
}

df_manual = pd.DataFrame(data)
print("Dataset Created using Pandas:\n")
print(df_manual)
```

```python
# 3) ii
# Loading from a CSV file (assuming you have data/students.csv)
# df_csv = pd.read_csv("data/students.csv")
# print("\nDataset Loaded from CSV:\n", df_csv.head())

# For demonstration, we’ll save and reload the above DataFrame
df_manual.to_csv("students.csv", index=False)
df_csv = pd.read_csv("students.csv")
print("\nLoaded from CSV file:\n", df_csv)
```

```python
# 3) iii

from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)
iris_df = iris.frame

print("\nIris Dataset Loaded from sklearn:")
print(iris_df.head())
```

```python
# 3 b 
import numpy as np
from statistics import mean, median, mode, variance, stdev

# Using the 'Marks' column from our DataFrame
marks = df_manual['Marks']

print("Mean:", marks.mean())
print("Median:", marks.median())
print("Mode:", marks.mode()[0])
print("Variance:", marks.var())
print("Standard Deviation:", marks.std())
```

```python
# 3) c i 

# Example: reshape using melt (long format)
reshaped = pd.melt(df_manual, id_vars=['Name'], value_vars=['Age', 'Marks'])
print("\nReshaped Data (using melt):\n", reshaped.head())
```

```python
# 3 c ii
# Example: Filter students with Marks > 85
filtered = df_manual[df_manual['Marks'] > 85]
print("\nFiltered Data (Marks > 85):\n", filtered)
```

```python
# 3 c iii - Merging two DataFrames
df_city = pd.DataFrame({
    'City': ['Bhopal', 'Delhi', 'Pune', 'Chennai'],
    'State': ['MP', 'Delhi', 'Maharashtra', 'TN']
})

merged_df = pd.merge(df_manual, df_city, on='City', how='left')
print("\nMerged DataFrame:\n", merged_df)
```

```python
# 3 c iv
# Create a DataFrame with missing values
df_missing = pd.DataFrame({
    'Name': ['Aman', 'Bhavna', 'Chetan', 'Diya'],
    'Marks': [85, None, 76, None]
})

print("\nOriginal DataFrame with Missing Values:\n", df_missing)

# Fill missing values with mean
df_filled = df_missing.fillna(df_missing['Marks'].mean())
print("\nAfter Filling Missing Values (with mean):\n", df_filled)

# Or drop missing values entirely
df_dropped = df_missing.dropna()
print("\nAfter Dropping Missing Values:\n", df_dropped)
```

```python
# 3 c v - Feature Normalization — Min-Max Normalization
# Using Marks column for normalization
marks = df_manual['Marks']
min_marks = marks.min()
max_marks = marks.max()

normalized_marks = (marks - min_marks) / (max_marks - min_marks)
df_manual['Normalized_Marks'] = normalized_marks

print("\nAfter Min-Max Normalization:\n", df_manual[['Name', 'Marks', 'Normalized_Marks']])
```
