```python
import math

# Example numbers
x = 7.8
y = 16

print("Number:", x)
print("math.floor(x):", math.floor(x))   # Largest integer <= x
print("math.ceil(x):", math.ceil(x))     # Smallest integer >= x
print("math.sqrt(y):", math.sqrt(y))     # Square root (float)
print("math.isqrt(y):", math.isqrt(y))   # Integer square root
print("math.gcd(12, 18):", math.gcd(12, 18))  # Greatest Common Divisor
print("math.factorial(5):", math.factorial(5))  # Factorial of 5
```

```python
import numpy as np

# Create a NumPy array
arr = np.array([[1, 2, 3], [4, 5, 6]])

print("Array:\n", arr)
print("ndim (number of dimensions):", arr.ndim)
print("shape (rows, cols):", arr.shape)
print("size (total elements):", arr.size)

# Common methods
print("Sum of all elements:", arr.sum())
print("Mean of all elements:", arr.mean())
print("Sorted array (flattened):", np.sort(arr, axis=None))
print("Sine of each element (in radians):\n", np.sin(arr))
```

```python
# For determinant and eigenvalues, use numpy or scipy
from numpy.linalg import det, eig

matrix = np.array([[2, 1],
                   [1, 2]])

print("Matrix:\n", matrix)
print("Determinant:", det(matrix))

eigenvalues, eigenvectors = eig(matrix)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

```python
# Original 1D list
lst = [1, 2, 3, 4, 5, 6, 7, 8, 9]

arr = np.array(lst)
print("Original 1D Array:", arr)

# Reshape into 2D (3 rows, 3 columns)
arr_2d = arr.reshape(3, 3)
print("\n2D Array (3x3):\n", arr_2d)

# Reshape into 3D (1 block, 3 rows, 3 columns)
arr_3d = arr.reshape(1, 3, 3)
print("\n3D Array (1x3x3):\n", arr_3d)
```

```python
zeros_matrix = np.zeros((3, 3))
ones_matrix = np.ones((2, 4))
identity_matrix = np.eye(3)  # Identity matrix

print("3x3 Zero Matrix:\n", zeros_matrix)
print("\n2x4 Ones Matrix:\n", ones_matrix)
print("\n3x3 Identity Matrix:\n", identity_matrix)
```

```python
from scipy import linalg

A = np.array([[4, 2],
              [3, 1]])

det_A = linalg.det(A)
print("Matrix A:\n", A)
print("Determinant of A (using SciPy):", det_A)
```

```python
# Eigenvalue & Eigenvector using SciPy
eigenvalues, eigenvectors = linalg.eig(A)

print("Eigenvalues (SciPy):", eigenvalues)
print("Eigenvectors (SciPy):\n", eigenvectors)
```
